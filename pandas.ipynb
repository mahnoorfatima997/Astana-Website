## Data analysis

### Subtask:
Analyze the distribution of key variables and relationships between price and other features in the cleaned Airbnb dataset.


**Reasoning**:
Analyze the distribution of price, relationship between price and room_type, price and minimum_nights, and the influence of other variables on price. Generate descriptive statistics, visualizations, and correlations as needed.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Price distribution analysis
print("Descriptive Statistics for Price:\n", df['price'].describe())
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# 2. Price vs. Room Type
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', data=df, palette="Set3")
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()

# Calculate average price for each room type
average_price_by_room_type = df.groupby('room type')['price'].mean()
print("\nAverage Price by Room Type:\n", average_price_by_room_type)


# 3. Price vs. Minimum Nights
correlation = df['price'].corr(df['minimum nights'])
print(f"\nCorrelation between Price and Minimum Nights: {correlation}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='minimum nights', y='price', data=df, alpha=0.5, color='green')
plt.title('Price vs. Minimum Nights')
plt.xlabel('Minimum Nights')
plt.ylabel('Price')
plt.show()

# 4. Influence of other variables on price
numerical_cols = ['number of reviews', 'availability 365', 'calculated host listings count']
for col in numerical_cols:
    correlation = df['price'].corr(df[col])
    print(f"\nCorrelation between Price and {col}: {correlation}")
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=col, y='price', data=df, alpha=0.5)
    plt.title(f'Price vs. {col}')
    plt.xlabel(col)
    plt.ylabel('Price')
    plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(x='neighbourhood group', y='price', data=df, palette="Set2")
plt.title('Price vs. Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()

average_price_by_neighbourhood = df.groupby('neighbourhood group')['price'].mean()
print("\nAverage Price by Neighbourhood Group:\n", average_price_by_neighbourhood)
```
